```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('DataCoSupplyChainDataset.csv',encoding='latin-1')
df.head(50)
```

```python
df.columns
```

```python
dfBase= df[['order date (DateOrders)','Order Item Total','Order Item Id','Product Card Id','Product Category Id',
            'Customer Id','Customer City','Customer State','Customer Country','Customer Segment',
            'Days for shipping (real)', 'Days for shipment (scheduled)','Delivery Status','shipping date (DateOrders)',
            'Product Price','Shipping Mode','Type','Order Item Quantity','Order Item Discount','Order Profit Per Order']]
dfBase.head(50)
```

```python
dfBase.info()
```

```python
dfBase.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
le = OrdinalEncoder()
dfBase[['Customer Segment']]  = le.fit_transform(dfBase[['Customer Segment']])
dfBase[['Customer Segment','Customer Id']]
```

```python
aggre = {'Order Item Id':'count','Order Item Total':'sum','Order Item Quantity':'sum','order date (DateOrders)':['max','min'],
         'Days for shipping (real)':'mean','Customer Segment':'max','Order Item Discount':'mean','Order Profit Per Order':'mean'}
aggregatedDict = dfBase.groupby('Customer Id').aggregate(aggre)
aggregatedDict
```

```python
aggregatedDict['order date (DateOrders)','max'] = pd.to_datetime(aggregatedDict['order date (DateOrders)','max'],format='mixed')
aggregatedDict['order date (DateOrders)','min'] = pd.to_datetime(aggregatedDict['order date (DateOrders)','min'],format='mixed')
aggregatedDict['order date (DateOrders)']
```

```python
aggregatedDict['Recency'] = (pd.to_datetime('2018-02-02') - aggregatedDict['order date (DateOrders)','max']).dt.days
aggregatedDict['Recency']
```

```python
aggregatedDict['Time since first purchase'] = (pd.to_datetime('2018-02-02') - aggregatedDict['order date (DateOrders)','min']).dt.days
```

```python
aggregatedDict = pd.DataFrame(aggregatedDict)
```

```python
aggregatedDict.columns
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(aggregatedDict['Order Item Total'])

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(aggregatedDict['Recency'])

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(aggregatedDict['Time since first purchase'])

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(aggregatedDict['Order Item Quantity'])

plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(aggregatedDict.corr(),annot= True)
```

```python
X = aggregatedDict.drop(['Order Item Total','order date (DateOrders)','Days for shipping (real)'],axis = 1)
X.head(50)
```

```python
Y = aggregatedDict['Order Item Total']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,train_size=0.7,random_state=9876)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_true = y_test
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2Score = r2_score(y_true,y_predict)
mae = mean_absolute_error(y_true,y_predict)
mse = mean_squared_error(y_true,y_predict)
rmse = np.sqrt(mse)
```

```python
print(f"R2 Score: {r2Score:.2f}")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f} and RMSE: {rmse:.2f}")
```
